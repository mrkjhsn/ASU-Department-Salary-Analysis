### In any given year, a number of ASU departments show an absence in activity, and a number of departments were either created or received funding which they did not have in the previous year.  Based on the names of departments that fall into each of these categories, I'm interested in the relationships I can discover.

```python
import pyodbc
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker
import seaborn as sns

# databaseConnection = 'DRIVER={SQL Server}; SERVER=ServerName; Database=DatbaseName; UID=UserId; PWD=password;'
connection = pyodbc.connect('DRIVER={SQL Server};SERVER=WINDOWS-A6S01I2\SQLEXPRESS;DATABASE=ASU_Professor_Salary;Trusted_Connection=yes;')
cursor = connection.cursor()

asu_dept_add_remove = '''
  SELECT *
  FROM [ASU_Professor_Salary].[dbo].[ASU_Dept_Changes]
  '''
# save the sql query to a pandas dataframe
asu_dept_add_remove = pd.read_sql(asu_dept_add_remove, connection)
connection.close()

asu_dept_add_remove.head()
```

```python
# in any given year, how many departments receive no funding, or receive funding after not receiving it in the previous year?
asu_dept_add_remove.groupby([asu_dept_add_remove.columns[0], asu_dept_add_remove.columns[2]])[asu_dept_add_remove.columns[0]].count()
```

```python
asu_dept_add_remove[asu_dept_add_remove['ADD_or_REM'] == 'REM'].head()
```

```python
asu_dept_add_remove['tokenize_text'] = asu_dept_add_remove['Department'].apply(lambda x: x.split(' '))

asu_dept_add_remove.head()
```

## Count tokens for department names that have been removed

```python
stopwords = ['university', 'center', 'office', 'systems', 'services', 'support', 'solutions', 'school'
            'student', 'academic', 'development', 'school','career', 'letters']

rem_dept_tokens = []

for row in asu_dept_add_remove[asu_dept_add_remove['ADD_or_REM'] == 'REM'].loc[:,'tokenize_text']:
    for department_name in row:
        if len(department_name) > 4 and department_name.lower() not in stopwords:
            rem_dept_tokens.append(department_name)

rem_dept_tokens_df = pd.DataFrame(rem_dept_tokens, columns=['tokens'])

rem_dept_tokens_count = rem_dept_tokens_df.groupby('tokens')['tokens'] \
    .count() \
    .reset_index(name= 'token_count') \
    .sort_values(by='token_count', ascending=False)

rem_dept_tokens_count['add_rem'] = 'rem'

rem_dept_tokens_count.sort_values(by='token_count', ascending=False).head(20)
```

## Count tokens for department names that have been added

```python
stopwords = ['university', 'center', 'office', 'systems', 'services', 'support', 'solutions', 'school'
            'student', 'academic', 'development', 'school','career', 'letters']

add_dept_tokens = []

for row in asu_dept_add_remove[asu_dept_add_remove['ADD_or_REM'] == 'ADD'].loc[:,'tokenize_text']:
    for department_name in row:
        if len(department_name) > 4 and department_name.lower() not in stopwords:
            add_dept_tokens.append(department_name)

add_dept_tokens_df = pd.DataFrame(add_dept_tokens, columns=['tokens'])

add_dept_tokens_count = add_dept_tokens_df.groupby('tokens')['tokens'] \
    .count() \
    .reset_index(name= 'token_count') \
    .sort_values(by='token_count', ascending=False)

add_dept_tokens_count['add_rem'] = 'add'

add_dept_tokens_count.sort_values(by='token_count', ascending=False).head(20)
```

```python
add_rem_tokens = rem_dept_tokens_count.sort_values(by='token_count', ascending=False).head(50) \
    .append(add_dept_tokens_count.sort_values(by='token_count', ascending=False).head(50), sort=True)

add_rem_tokens = pd.melt(add_rem_tokens, id_vars=['tokens', 'token_count'])

add_rem_tokens.head()

# add_rem_tokens.groupby('tokens').count()
```

```python
plt.subplots(figsize=(15,30))

sns.barplot(x=add_rem_tokens.loc[:,'token_count'], \
             y=add_rem_tokens.loc[:,'tokens'], \
             hue=add_rem_tokens.loc[:,'value'], \
             data=add_rem_tokens)

plt.title('ASU Department Keyword Add/Remove Activity',fontsize=20)
plt.ylabel('Department Name Tokens',fontsize=17)
plt.xlabel('Count of Tokens',fontsize=17)
plt.tick_params(labelsize=13)
plt.legend(loc='center right',fontsize=15)
```
